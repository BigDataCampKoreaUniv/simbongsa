```python

koParser = nltk.RegexpParser("""
    NP: {<N.*>*<N.*>?}
    OP: {<N.*>*<JKO>*}
    VP: {<M.*><V.*>*<E.*>*}
""")
```

```python

def parsing_contents (data) : 
    dics = {}
    errorid =[]
    for i in range(len(spec_up)):
        lists = []
        try: #활동 내용 없으면 에러나서 
            koTags = kkma.pos(data["활동내용"][i])
            tree = koParser.parse(koTags)
        except : 
            errorid.append(i)
        for subtree in tree.subtrees():
            if subtree.label() == "NP":
                a = " ".join(e[0] for e in list(subtree))
                #print(a)

                if a.find(' ') != -1:
                    for j in range(len(a.split())):
                        lists.append(a.split()[j])
                else :
                    lists.append(a)
                dics[i] = lists
    return [dics , errorid]
```

```python
def parsing_volunteer(data):
    dics = {}
    errorid =[]
    for i in range(len(spec_up)):
        lists = []
        try: #활동 내용 없으면 에러나서 
            koTags1 = kkma.pos(data["모집 대상"][i])
            koTags2 = kkma.pos(data["제목"][i])
            tree1 = koParser.parse(koTags1)
            tree2 = koParser.parse(koTags2)
        except : 
            pass
        for subtree1 in tree1.subtrees():
            if subtree1.label() == "NP":
                a = " ".join(e[0] for e in list(subtree1))
                #print(a)

                if a.find(' ') != -1:
                    for j in range(len(a.split())):
                        lists.append(a.split()[j])
                else :
                    lists.append(a)
                dics[i] = lists
        for subtree2 in tree2.subtrees():
            if subtree2.label() == "NP":
                a = " ".join(e[0] for e in list(subtree2))
                #print(a)

                if a.find(' ') != -1:
                    for j in range(len(a.split())):
                        lists.append(a.split()[j])
                else :
                    lists.append(a)
                dics[i] = lists
    return [dics , errorid]
```

```python
def makingMatrix(voca, dics):

    Matrix = [[0 for y in range(len(voca))] for x in range(len(dics))]
    key_voca = list(voca.keys())
    #print(key_voca)

    for i in range(len(key_voca)):
        for j in range(len(voca[key_voca[i]])):
            for k in range(len(dics)):
                for l in range(len(dics[k])):
                               if voca[key_voca[i]][j] in dics[k][l]:
                                   # print('{0} (은)는 {1} 에 속합니다'.format(dics[k][l],key_voca[i]))
                                    #print('그러므로 {0} 번째 게시글은 key_voca의 {1} 번째에 속합니다.\n'.format(k, i))
                                    Matrix[k][i]+=1
                        
    return Matrix
```

```python
def sum_ma(Matrix):
    sum_ma_li = []
    for i in Matrix :
        sum_ma = 0
        for j in i :
            sum_ma += j
        sum_ma_li.append(sum_ma)
    return sum_ma_li
```
